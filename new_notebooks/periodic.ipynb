```python
import matplotlib.pyplot as plt
import json
import matplotlib.patches as patches
import matplotlib.collections
import glob
import itertools
```

```python
files = sorted(glob.glob("../equil*"))
files
```

```python
i = 0
```

```python
for i in range(len(files)):
    with open(files[i], "r") as f:
        data = json.load(f)
    i += 1
        

    diff = [0, -3.5, +3.5]
    combs = list(itertools.product(diff, diff))

    fig, ax = plt.subplots()


    for (x, y, z), t in zip(data["x"], data["types"]):
        for dx, dy in combs:
            ax.add_patch(plt.Circle((x+dx, y+dy), data["sigmas"][int(t)], alpha=.5))
    plt.xlim(-1.75, 1.75)
    plt.ylim(-1.75, 1.75)
```
